```python
from flask import Flask, request, jsonify

app = Flask(__name__)

# Define a route to handle incoming requests
@app.route('/transactions', methods=['GET'])
def get_transactions():
    account_number = request.args.get('account_number')
    
    if account_number is None:
        return jsonify({'error': 'Please provide account_number parameter'}), 400
    
    # Filter transactions where the given account number is either the sender or receiver
    transactions_from_account = df[df['From Bank'] == int(account_number)].to_dict(orient='records')
    transactions_to_account = df[df['To Bank'] == int(account_number)].to_dict(orient='records')
    
    # Extract relevant information from transactions
    transactions_from_account = [{'Timestamp': txn['Timestamp'], 'Amount': txn['Amount Paid']} for txn in transactions_from_account]
    transactions_to_account = [{'Timestamp': txn['Timestamp'], 'Amount': txn['Amount Received']} for txn in transactions_to_account]
    
    return jsonify({'transactions_from_account': transactions_from_account, 'transactions_to_account': transactions_to_account})

if __name__ == '__main__':
    app.run(debug=True)
```
